```python
import os
import pandas as pd
```

```python
data_path = r'C:\Projects\connecteddatahub\data\gpt_dataframes'
```

```python
csv_files = [
    os.path.join(data_path, fname)
    for fname in os.listdir(data_path)
    if fname.lower().endswith('.csv')
]

# 2) Read each into a DataFrame
dfs = []
for csv_file in csv_files:
    try:
        df = pd.read_csv(csv_file)
        dfs.append(df)
    except Exception as e:
        print(f"⚠️ Skipping {csv_file}: {e}")

if not dfs:
    raise RuntimeError(f"No CSV files read from {data_path}")

# 3) Concatenate all DataFrames
all_df = pd.concat(dfs, ignore_index=True)

# 4) Count unique occurrences in the 'Name' column
name_counts = (
    all_df['Name']
    .value_counts(dropna=False)       # include NaNs if you want to count them
    .rename_axis('Name')
    .reset_index(name='Count')
)

# 5) (Optional) sort by count descending
name_counts = name_counts.sort_values(by='Count', ascending=False)

# 6) Output
print(name_counts)
```
