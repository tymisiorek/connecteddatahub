```python
import os
import pandas as pd
import numpy as np
```

```python
board_path = r'C:\Users\tykun\OneDrive\Documents\SchoolDocs\VSCodeProjects\connectedData\board_analysis\board_dataframes'
```

```python
years = ["1999","2000","2005","2007","2008","2009","2010","2011","2013","2018",]
```

```python
year_to_institutions = {}
for year in years:
    file_path = os.path.join(board_path, f"{year}_boards.csv")
    
    if os.path.exists(file_path):
        df = pd.read_csv(file_path)
        df = df[df['PrimarySample'] == True]
        institutions = set(df['Institution'].dropna().astype(str))
        year_to_institutions[year] = institutions
        
```

```python
all_institutions = set.union(*year_to_institutions.values())

# Step 3: Find institutions not in every year
not_in_every_year = {
    inst: [year for year, insts in year_to_institutions.items() if inst in insts]
    for inst in all_institutions
    if sum(inst in insts for insts in year_to_institutions.values()) < len(years)
}

# Step 4: Output
print(f"\n{len(not_in_every_year)} institutions did not appear in every year:")
for inst, present_years in sorted(not_in_every_year.items()):
    print(f"- {inst}: appears in {present_years}")
```
