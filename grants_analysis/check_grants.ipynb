```python
import pandas as pd
import os
```

```python
data_path = r'C:\Projects\connecteddatahub\data'
grants_path = os.path.join(data_path, 'grants', 'cleaned_grants_2010_2024.csv')
```

```python
grants_df_full = pd.read_csv(grants_path)
```

```python
print(grants_df_full.shape)
print(grants_df_full.columns)
```

```python
print(grants_df_full['year'].value_counts().sort_index())
```

```python
grants_df = grants_df_full[(grants_df_full['year'] < 2019) & (grants_df_full['year'] >= 2010)]
```

```python
print(grants_df['year'].value_counts().sort_index())
```

```python
print(
    grants_df.loc[
        (grants_df['year'] == 2013) & (grants_df['recip_name'].str.contains("university", case=False, na=False)),
        'recip_name'
    ].value_counts().head(20)
)
```

```python
# Get top 20 university recipients in 2013
top_20_2013 = (
    grants_df.loc[
        (grants_df['year'] == 2013)
        & (grants_df['recip_name'].str.contains("university", case=False, na=False)),
        'recip_name'
    ]
    .value_counts()
    .head(20)
    .index
)

# Show yearly counts for those same recipients
print(
    grants_df.loc[
        grants_df['recip_name'].isin(top_20_2013)
    ]
    .groupby(['year', 'recip_name'])
    .size()
    .unstack(fill_value=0)
)
```

```python
grants_df.to_csv(os.path.join(data_path, 'grants', 'cleaned_grants_2010_2018.csv'), index = False)
```
